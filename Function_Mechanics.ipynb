## SCOPE (LEGB RULE)
LEGB = Local → Enclosing → Global → Built-in

Think of Python as searching for variable values in layers, from smallest to biggest.

Let’s understand each layer:

1. Local (L)
- Variables created inside the current function.

2. Enclosing (E)
- Variables from outer functions (used in closures).

3. Global (G)
- Variables defined at top level in the script.

4. Built-in (B)
- Keywords and built-ins like len, print, range.

```python
x=10 # Global
def outer():
    x=20 # Enclosing
    def inner():
        x=30 # Local
        print(x)
    inner()
    print(x)
outer()
print(x)
```

## PASS-BY-OBJECT-REFERENCE

Python is not pass-by-value
Python is not pass-by-reference
Python uses pass-by-object-reference (also called call-by-sharing).

Meaning:
>  A function receives a reference to the object, not the object itself.
Whether the object changes depends on whether it is mutable or immutable.

Mutable vs Immutable Behavior
Immutable Types
- int
- float
- str
- tuple
- bool

You cannot modify the original object.

```python
def modify(n:int):
    n+=10

x=10
modify(x)
print(x)
```

**Reason: n += 10 creates a new integer object**

```python
def modify(s:str):
    s.lower()

name='KABIL'
modify(name)
print(name)
```

**Reason: s.lower() creates a new string object**

Mutable Types
- list
- dict
- set
- custom objects

You can modify the original object through the reference.

```python
def modify(ls:list):
    ls.append(5)

ns=[1,2,3,4]
modify(ns)
print(ns)
```

```python
def modify(d:dict):
    d['friend']='Kabishan'

dic={'name':'Kabileshwaran','age':23}
modify(dic)
print(dic)
```

## CLOSURES

A closure occurs when a function remembers and uses variables from an outer function even after the outer function has finished executing.

This is a core concept behind:
- decorators
- function factories
- event handlers
- dynamic configuration
- memoization

```python
def outer(msg):
    def inner():
        print(msg)
    return inner

f = outer("Hello")
f()  # prints "Hello"
```
